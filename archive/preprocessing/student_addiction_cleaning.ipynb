```python
import pandas as pd

sad_train = pd.read_csv('raw_data/student-addiction-train-raw.csv')
sad_test = pd.read_csv('raw_data/student-addiction-test-raw.csv')
```

```python
print(sad_train.shape)
print(sad_test.shape)
```

```python
# drop NULLs
for df in [sad_train, sad_test]:
    df.dropna(inplace=True)
    df.reset_index(drop=True, inplace=True)
    
    print(df.shape)
```

```python
sad_train.sample(15)
```

```python
# Map all variables from Yes/No to 1/0
# Note that all variables are binary

for df in [sad_train, sad_test]:
    for col in df.columns[0:]:
        df[col] = df[col].map({'Yes': 1, 'No': 0})
        df[col] = df[col].astype(int)
```

```python
sad_train.sample(15)
```

```python

# save dataset two folders up from current folder and then from there in src/data/clean/

sad_train.to_csv('../../src/data/clean/student-addiction-train.csv', index=False)
sad_test.to_csv('../../src/data/clean/student-addiction-test.csv', index=False)
```
